[1390. Four Divisors](https://leetcode.com/problems/four-divisors/description/?envType=daily-question&envId=2026-01-04)

```python
from typing import List
from sympy import divisors

class Solution:
    def sumFourDivisors(self, nums: List[int]) -> int:
        result = 0
        for _, v in enumerate(nums):
            divs = divisors(v) # using sympy library
            if len(divs) == 4:
                result += sum(divs)
        return result
```

```python
s = Solution()

s.sumFourDivisors([21,21])
```

```python
from typing import List

class Solution2:
    def sumFourDivisors(self, nums: List[int]) -> int:
        result = 0
        for _, v in enumerate(nums):
            divs = self.divisors(v) # using sympy library
            if len(divs) == 4:
                result += sum(divs)
        return result

    def divisors(self, n: int) -> List[int]:
        divisors = []
        # Only iterate up to sqrt(n)
        for i in range(1, int(n**0.5) + 1):
            if n % i == 0:
                divisors.append(i)
                if i != n // i:  # Avoid duplicate for perfect squares
                    divisors.append(n // i)
            if len(divisors) > 4: # stop calculations
                return divisors
        return divisors
```
