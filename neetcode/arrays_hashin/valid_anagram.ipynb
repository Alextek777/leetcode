[Valid Anagram](https://neetcode.io/problems/is-anagram/question?list=neetcode150)

```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False

        cache = {}
        for i in range(0, len(s)):
            cache[s[i]] = cache.get(s[i], 0) + 1
            cache[t[i]] = cache.get(t[i], 0) - 1

        for _, v in cache.items():
            if v != 0:
                return False
        return True
```

```python
s = Solution()

s.isAnagram("123", "321")
```
