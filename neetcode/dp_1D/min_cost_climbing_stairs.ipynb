[Min Cost Climbing Stairs](https://neetcode.io/problems/min-cost-climbing-stairs/question?list=neetcode150)

BFS with memorization

```python
from typing import List

class Solution:
    def minCostClimbingStairs(self, cost: List[int]) -> int:
        self.cache = {}
        return min(
            self.minCost(cost, 0), 
            self.minCost(cost, 1), 
        )
        
    def minCost(self, cost: List[int], i: int) -> int:
        if i >= len(cost):
            return 0
        
        if i in self.cache:
            return self.cache[i]

        curCost = cost[i] + min(
            self.minCost(cost, i+1), 
            self.minCost(cost, i+2), 
        )

        self.cache[i] = curCost

        return curCost
```

```python
s = Solution()
s.minCostClimbingStairs([1,2,1,2,1,1,1])
```

```python
s.minCostClimbingStairs([1,2,3])
```
