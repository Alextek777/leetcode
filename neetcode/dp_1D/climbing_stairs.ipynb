[Climbing Stairs](https://neetcode.io/problems/climbing-stairs/question?list=neetcode150)

```python
class Solution:
    def __init__(self):
        self.cache = {}
        
    def climbStairs(self, n: int) -> int:
        if n == 0:
            return 1
        if n < 0:
            return 0
            
        if n in self.cache:
            return self.cache[n]

        result = self.climbStairs(n - 1) + self.climbStairs(n - 2)
        self.cache[n] = result

        return result
```

```python
s = Solution()

s.climbStairs(50)
```

```python
class Solution2:        
    def climbStairs(self, n: int) -> int:
        one, two = 1, 1
        for _ in range(n-1):
            one, two = two, one + two
        return two
```

```python
s2 = Solution2()
s2.climbStairs(50)
```
