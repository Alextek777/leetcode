[1143. Longest Common Subsequence](https://leetcode.com/problems/longest-common-subsequence/solutions/7482713/longest-common-subsequence-recursion-bot-qfr0/)

### Approach
We use bottom-up DP.

### DP Definition:
#### Let:
```
dp[i][j] = length of the LCS between
s1[0..i-1] and s2[0..j-1]
```

### Transitions
If characters match:

```dp[i][j]=1+dp[i−1][j−1]```

### Otherwise:

```dp[i][j]=max(dp[i−1][j],dp[i][j−1])```

### Base Case

```
dp[0][*] = dp[*][0] = 0
(An empty string has LCS length 0 with any string)
```

### Result

```
The final answer is stored in:
dp[m][n]
where m and n are the lengths of the two strings.
```

![lcs.png](lcs.png)

````python
class Solution:
    def longestCommonSubsequence(self, text1: str, text2: str) -> int:
        m, n = len(text1), len(text2)

        dp = [[0 for _ in range(n+1)] for _ in range(m+1)]

        for i in range(1, m+1):
            for j in range(1, n+1):
                if text1[i-1] == text2[j-1]:
                    dp[i][j] = 1 + dp[i-1][j-1]
                else:
                    dp[i][j] = max(dp[i-1][j], dp[i][j-1])

        return dp[m][n]
````

````python
s = Solution()

s.longestCommonSubsequence("12","123")
````
